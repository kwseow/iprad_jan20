## Reshape

You are able to change the dimension of an array using reshape()

```
narr3 = np.arange(24)
print('Original Array')
print(narr3)
```

```
narr4 = narr3.reshape(2,3,4)
print("Reshapre the array by 2 * 3 * 4 Dimension array")
print(narr4)
```

```
narr4 = narr3.reshape(4,6)
print("Reshapre the array by 4 * 6 Dimension array")
print(narr4)
```

It is important to note when reshaping the array, the number of elements must ‘fit’ exactly to the dimension.

For example, if narr1 = [1,2,3,4,5,6,7,8], there are 8 elements in the narr1 array, hence it will not be able to fit in a 2 X 3 array   

It is like putting 8 soccer balls in 6 small boxes.

## Concatenation
Concatenation, or joining of two arrays in NumPy, is primarily accomplished through the routines <b>np.concatenate, np.vstack, and np.hstack</b>. np.concatenate takes a tuple or list of arrays as its first argument, as we can see here:

```
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

You can also concatenate more than two arrays at once:

```
z = [99, 99, 99]
print(np.concatenate([x, y, z]))
```

<b>np.concatenate</b> can also be used for two-dimensional arrays:

```
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
```

```
# concatenate along the first axis 
np.concatenate([grid, grid])
```

```
# concatenate along the second axis (zero-indexed) 
np.concatenate([grid, grid], axis=1)
```

For working with arrays of mixed dimensions, it can be clearer to use the <b>np.vstack</b> (vertical stack) and <b>np.hstack</b> (horizontal stack) functions:

```
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],
                 [6, 5, 4]]) 
# vertically stack the arrays
np.vstack([x, grid])
```

```
# horizontally stack the arrays 
y = np.array([[99],
              [99]])
np.hstack([grid, y])
```

Similarly, np.dstack will stack arrays along the third axis.

#### Splitting of arrays
The opposite of concatenation is splitting, which is implemented by the functions <b>np.split</b>, <b>np.hsplit</b>, and <b>np.vsplit</b>. For each of these, we can pass a list of indices giving the split points:

```
x=[1,2,3,99,99,3,2,1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

Notice that N split points lead to N + 1 subarrays. The related functions np.hsplit and np.vsplit are similar:

```
grid = np.arange(16).reshape((4, 4))
grid
```

```
upper, lower = np.vsplit(grid, [2]) 
print(upper)
print(lower)
```

Similarly, np.dsplit will split arrays along the third axis.

### Basic slicing
This is an operation to extract a subset of the array.  There maybe a need to extract a subset for detail analysis or modifying of the data.  For example, modifying a single value in an array


```
x1 = np.arange(10)
print(x1)
```

```
x1[0]=10
print(x1[0])
```

**Slice** function that allow the retrieving  from a starting and end array index at a pre-determine increment steps

slice(start, end, increment) 

image.png


```
x2 = np.arange(30)
x2a = slice(3,10,2)
print("Slicing X2: ", x2[x2a])
```

An Array can be sliced without the use of the Slice function 
The same result can also be obtained by giving the slicing parameters separated by a **colon** : (start:stop:step) directly to the **ndarray** object.

```
x2[3:10:2]
```

Array can be sliced from specific index

```
x3 = np.arange(20)
print("Array x3 :")
print(x3)
```

```
print("Array Sliced from position 2")
print(x3[2:])
```

## Statistical Functions

NumPy provides the following Statistical Functions
* numpy.amin() and numpy.amax() (Minimum and Maximum) 
* numpy.ptp()
* numpy.percentile()
* numpy.median()
* numpy.mean()
* numpy.average()
* Standard Deviation
* Variance 

Most of the statistical functions have the common syntax. 

**amin() & amax()**

These functions determines the minimum and maximum based on the specific axis
* Axis = 0 refer to the vertical
* Axis = 1 refer to the hortizontal

This is applicable to other functions


```
a = np.array([[3,7,5],[8,4,3],[2,4,9]])
print ('Our array is:')
print (a)

print ('Applying amin(a,axis=1) function:')
print (np.amin(a,axis=1))

print ('Applying amin(a,axis=0) function again:')
print (np.amin(a,0))

print ('Applying amax(a) function:')
print (np.amin(a))
```

**Percentile**

Compute the q-th percentile of the data along the specified axis.

Returns the q-th percentile(s) of the array elements.



```
a = np.array([[30,40,70],[80,20,10],[55,90, 35],[55,90,60]])
print (np.percentile(a,50))
print (np.percentile(a,50, axis=1))
print (np.percentile(a,50, axis=0))
```
