### Creating Series Data Structure

```
import pandas as pd
import numpy as np

#Create from ndarray
data = np.array(['a', 'b', 'c', 'd', 'e'])
s = pd.Series(data)
print('Series s created from ndarray \n', s)
```

```
#Create from ndarray with label
t = pd.Series(data, index=[100,101,102,103,104])
print('Series t created from ndarray with label \n', t)
```

```
#Create From dict 
data2 = {'a':0, 'b':1.5, 'c':2.5}
d = pd.Series(data2)
print('Sereies d created from dict \n', d)
```

## Accessing Elements from a Series Data Structure

```
#This is to create the Series for explaining accessing the data
s= pd.Series(np.arange(1,9), index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
print(s)

#Access by the integer index
print('Element at position 1 is : ', s[1])
```

```
#Access a range of elements
print('Elements from postion 1 to 5: \n ', s[1:5])
```

```
#Access the first 5 elements
print('First 5 elements \n', s[:5])
```

```
#Accest the last 3 elements
print('Last 3 elements \n', s[-3:])
```

```
#Access the element by label(index)
print('Eements by Label \n', s['c'])

print('Eements by Label \n', s[['a', 'c']])
```

### Data Frame Data Structure

DataFrame are:
* Two Dimensional labelled respectively
* Data can be heterogeneous
* Size and data is mutable
* Able to perform Arithmetic operations on rows and columns

image.png


Data Frame can be created from the following inputs
* Lists
* Dict
* Series
* Ndarray
* Another Data Frame


```
#Create from a list of data
listData = [1, 3,5,6,8, 10]
listDF = pd.DataFrame(listData)
print('Created from List\n')
print(listDF)
```

```
print('Created from List')
data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'])
print(df)
```

```
print('Created from List')
print('Define the data type for age \n')
data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'],dtype=float)
print(df)
```

```
print('Created from ndarray')
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'], 'Age':[28,34,29,42]}
df = pd.DataFrame(data,columns=['Name','Age'],dtype=float)
print(df)
```

```
print('Created from List')
print('This example is created as an indexed(labelled DataFrame \n')
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df = pd.DataFrame(data, index=['rank1','rank2','rank3','rank4'])
print(df)
```

```
print('Created from List of Dict')
data = [{'a': 1, 'b': 2}, {'a': 5, 'b': 10, 'c': 20}]
df = pd.DataFrame(data)
print(df)
```

## Row Selection and Slicing

The row from the Dataframe can be retrieved by label or integer
* Use loc  to retrieve by label
* Use iloc to retrieve by integer

```
d = {'one' : pd.Series([1, 4, 3,4], index=['a', 'b', 'c','e']),
    'two' : pd.Series([1, 5, 3, 4, 8], index=['a', 'b', 'c', 'd','e']),
    'three' : pd.Series([5, 15, 13, 4,9], index=['a', 'b', 'c', 'd','f'])}
df = pd.DataFrame(d)
# Original Data Frame
print('Original Data Frame')
df.head()
```

```
# Selection of Row by Labels
print('Selection of Row b')
print (df.loc['b'])
```

```
#Selection of Row by range of label
print('Selection of Range of Row ')
print (df.loc['b':'d'])
```

```
# Selection of Row by Integer
print('Selection of Row 3 ')
print (df.iloc[3])
```

```
# Selection of Row by range of Integer
print('Selection of Range of Row')
print (df.iloc[3:5])
```

## Sorting

There maybe incidents where there is a need to sort the data 
E.g. Analyzing and training certain range of data.
There are two sorting available in Pandas:
* By Label
* By Actual Value 

The array can by sorted by the **label**(index) using **sort_index**()


```
# Create the unsorted data
unsorted_data=pd.DataFrame({'Col1':[5, 10,3, 8, 9 ], 'Col2':[3,4,2,5, 9], 'Col3': [4,0,5, 3,2] }, index=[4, 2, 5,1,6] )
print('Unsorted Data')
unsorted_data
```

```
# Sort the data by Label
sort_byLabel = unsorted_data.sort_index()
print('Sorting by label (index)')
sort_byLabel
```

```
# Order of Sorting by Label
sort_byLabel2 = unsorted_data.sort_index(ascending=False)
print('Sorting by label (index) in descending order')
sort_byLabel2
```

**Sorting by Values**

The array can by sorted by the **label**(index) using **sort_value**(). It accept a **‘by’** argument to determine which column to sort

The array can be sorted by multiple column. For example, sort_value(by=[‘col1’, ‘col2’]), the array will be sorted by col1 followed by col2


```
unsorted_df= pd.DataFrame({'col1':[2,4,1,3],'col2':[1,6,3,7],
                           'col3':[6,1,3,2], 'col4':[3,5,2,4]})
# Original Unsorted Array
print('Original Unsorted Array')
unsorted_df
```

```
# Sorting by Col1
print('Sort the array by Col1')
sorted_df=unsorted_df.sort_values(by='col1')
sorted_df
```

```
# Sorting by Col1 and cCol2
print('Sort the array by Col1 then by Col2')
sorted_df=unsorted_df.sort_values(by=['col1','col2'])
sorted_df
```

**Sorting using Algorithm**

**sort_values**() has a provision to choose the algorithm from mergesort, heapsort and quicksort


```
# Using Megasort
sorted_df=unsorted_df.sort_values(by='col1' ,kind='mergesort')
print('Using Megasort')
sorted_df
```
