## Scenario 2

```
file_path = "/content/drive/My Drive/Colab Notebooks/IPRAD/dirty_data.csv"
df = pd.read_csv(file_path)
```

```
#Create a duplicate for dataframe df
df_Ex2 = df.copy()
df_Ex2.head()
```

### Rule 1
Remove all rows that have non-numeric values

```
df_Ex2 = df_Ex2.drop(df_Ex2[df_Ex2['floor_size'].str.isalpha()== True].index)
```

```
df_Ex2.head()
```

**Add the codes to drop all non-alphanumeric characters**


```
df_Ex2 = df_Ex2.drop(df_Ex2[df_Ex2['households'].str.isalpha()== True].index)
df_Ex2 = df_Ex2.drop(df_Ex2[df_Ex2['median_income'].str.isalpha()== True].index)
df_Ex2 = df_Ex2.drop(df_Ex2[df_Ex2['median_house_value'].str.isalpha()== True].index)
```

```
df_Ex2.head()
```

The following code replaces black values with Nan

```
df_Ex2['floor_size'] = pd.to_numeric(df_Ex2['floor_size'], errors="coerce")
```

```
df_Ex2.head()
```

**Add the code where blank values are nullify**


```
df_Ex2['households'] = pd.to_numeric(df_Ex2['households'], errors="coerce")
df_Ex2['median_income'] = pd.to_numeric(df_Ex2['median_income'], errors="coerce")
df_Ex2['median_house_value'] = pd.to_numeric(df_Ex2['median_house_value'], errors="coerce")
```

```
df_Ex2.head()
```

### Rule 2
The following code will change null value to the mean value of the column

```
df_mean=df_Ex2.mean()
df_Ex2.loc[df_Ex2['floor_size'].isnull(),'floor_size'] = df_mean['floor_size']
```

```
df_Ex2.head()
```

**Add the code to change the NaN values to the means of the respective columns**

```
df_Ex2.loc[df_Ex2['households'].isnull(),'households'] = df_mean['households']
df_Ex2.loc[df_Ex2['median_income'].isnull(),'median_income'] = df_mean['median_income']
df_Ex2.loc[df_Ex2['median_house_value'].isnull(),'median_house_value'] = df_mean['median_house_value']
```
